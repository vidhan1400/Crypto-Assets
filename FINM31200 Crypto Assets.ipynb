$\large\textbf{2 }\quad\large\textbf{Import Libraries}$

```python
import numpy as np
import pandas as pd
import nasdaqdatalink
import scipy.stats as ss
import seaborn as sns
from functools import lru_cache
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
import datetime
import pandas_datareader as pdr
import warnings
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [15,10]
warnings.filterwarnings('ignore')
```

```python
def fetch_quandl(secs,start_date,end_date,columns):
    # Retrieve data using quandl.get_table
    nasdaqdatalink.ApiConfig.api_key = 'Your_API_key'
    data = nasdaqdatalink.get_table('QUOTEMEDIA/PRICES', ticker = secs, qopts={"columns":["ticker","date","adj_close","adj_volume"]}, date = { 'gte': start_date, 'lte': end_date })    
    return data
```

$\textbf{Get EOD (end of day) data for Traditional ETFs from Quandl}$

```python
secs = ['BIL','BNDX','COW','DBA','EMB',
        'FM','GLD','VWO','HYG','LQD',
        'MBB','PSP','SCHB','SLV','TIP',
        'UNG','USO','VEA','VNQ','SHY',
       'IEI', 'TLT']
start_date,end_date = '2017-11-15','2023-11-15'
etf_data = pd.DataFrame()
for i in range(len(secs)):
    if i ==0: 
        etf_data = fetch_quandl(secs[i],start_date,end_date,('adj_close','adj_volume'))
    
    else:
        etf_data = pd.concat([etf_data, fetch_quandl(secs[i],start_date,end_date,('adj_close','adj_volume'))], axis = 0, join = 'inner')
etf_data
```

```python
df = etf_data[etf_data['ticker']=='BIL'].set_index('date')
```

```python
fig, axs = plt.subplots(22, 1, figsize=(20, 75))

for i,sec in enumerate(secs):
    df = etf_data[etf_data['ticker']==sec]
    axs[i].plot(df['date'], df['adj_close'], label=sec)
    axs[i].set_title(f'Historical prices ' + str(sec),fontsize=15)
    axs[i].grid(linestyle='solid',color='grey', linewidth=0.6)
    axs[i].set_xlabel('Time',fontsize = 10)
    axs[i].legend(fontsize = 12)
plt.tight_layout()
plt.show()
```
