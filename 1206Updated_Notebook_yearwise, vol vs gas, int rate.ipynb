```python
def fetch_quandl(secs,start_date,end_date,columns):
    # Retrieve data using quandl.get_table
    nasdaqdatalink.ApiConfig.api_key = 'K2vNRYdRHsP6W8cEX-6q'
    data = nasdaqdatalink.get_table('QUOTEMEDIA/PRICES', ticker = secs, qopts={"columns":["ticker","date","adj_close","adj_volume"]}, date = { 'gte': start_date, 'lte': end_date })    
    return data
```

$\textbf{Get EOD (end of day) data for Traditional ETFs from Quandl}$

```python
secs = ['BIL','BNDX','COW','DBA','EMB',
        'FM','GLD','VWO','HYG','LQD',
        'MBB','PSP','SCHB','SLV','TIP',
        'UNG','USO','VEA','VNQ','SHY',
       'IEI', 'TLT','SPY']
start_date,end_date = '2017-11-15','2023-11-15'
etf_data = pd.DataFrame()
for i in range(len(secs)):
    if i == 0: 
        etf_data = fetch_quandl(secs[i],start_date,end_date,('adj_close','adj_volume'))
    else:
        etf_data = pd.concat([etf_data, fetch_quandl(secs[i],start_date,end_date,('adj_close','adj_volume'))], axis = 0, join = 'inner')
etf_data
```

## Hypothesis on Correlations Between BTC and Different Asset Classes' ETFS

- Bonds (BIL, BNDX, LQD, TIP, IEI, TLT): Cryptocurrencies like Bitcoin have shown periods of both positive and negative correlations with bonds. During times of market uncertainty or economic instability, some investors view cryptocurrencies as a hedge against traditional assets like bonds, resulting in potentially higher correlations. However, this correlation can vary based on market sentiment and risk perception.

- Commodities (COW, DBA, GLD, SLV, UNG, USO): Cryptocurrencies, particularly Bitcoin, have at times shown correlations with commodities like gold and silver. Some investors perceive cryptocurrencies as digital commodities or "digital gold," resulting in periods of positive correlation, especially during market turmoil or inflation concerns.

- Equities (FM, VWO, SCHB, VEA, VNQ): Cryptocurrencies have displayed varying correlations with global equities. During certain periods, they've shown a positive correlation, especially with emerging market stocks. However, this correlation can be less consistent due to the different market dynamics and regulatory environments in various regions.

- Fixed Income and High Yield (HYG, MBB, SHY): Cryptocurrencies have had mixed correlations with high-yield bonds and short-term treasuries. Their correlations with fixed income assets can fluctuate based on market sentiment, interest rate expectations, and macroeconomic factors.

- Diversified Portfolios (PSP): This ETF represents a diversified portfolio of global listed private equity. The correlation with cryptocurrencies might vary widely due to the nature of private equity investments, which can be less influenced by short-term market trends.

```python
# Calculate yearly correlations
yearly_correlations = {}

for year, group in bitcoin_price['Daily_Return'].groupby(bitcoin_price['Daily_Return'].index.year):
    yearly_data = pivot_data.loc[str(year)]  # Data for the current year
    yearly_correlations[year] = yearly_data.corrwith(group)

# Create a DataFrame for yearly correlations
yearly_correlations_df = pd.DataFrame(yearly_correlations)

# Display the DataFrame with yearly correlations
print(yearly_correlations_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have the DataFrame 'yearly_correlations_df' containing yearly correlations

# Create a heatmap for yearly correlations
plt.figure(figsize=(10, 8))
sns.heatmap(yearly_correlations_df, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Yearly Correlations: Bitcoin vs ETFs')
plt.xlabel('ETFs')
plt.ylabel('Year')
plt.show()
```

```python
# Number of subplots to create
num_subplots = len(yearly_correlations_df.index)

# Create subplots for each ETF's yearly correlation trend with Bitcoin
fig, axes = plt.subplots(num_subplots, 1, figsize=(8, 6*num_subplots), sharex=True)

for i, ETF in enumerate(yearly_correlations_df.index):
    axes[i].plot(yearly_correlations_df.loc[ETF])
    axes[i].set_title(f'Yearly Correlation Trend: Bitcoin vs {ETF}')
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Correlation')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Loop through each ETF in the index and plot its yearly correlations
for ETF in bond_etfs:
    plt.plot(yearly_correlations_df.loc[ETF], label=ETF)

plt.xlabel('Year')
plt.ylabel('Correlation')
plt.title('Yearly Correlation Trends: Bitcoin vs ETFs')
plt.legend()
plt.grid(True)
plt.show()
```

## Volume Correlation with Transation Fee Core Findings:

1. Positive Correlation Between Volume and Fee: There is a clear positive correlation between transaction volume and transaction fees. This indicates that as more transactions occur within the Bitcoin network, the total transaction fees tend to increase.


2. Higher Correlation in USD Terms: The correlation is stronger when transaction volumes and fees are measured in USD rather than in raw Bitcoin. This can be attributed to the influence of Bitcoin's price volatility. When measuring in Bitcoin, the inherent price fluctuations can mask the underlying relationship between transaction volume and fees. In contrast, measuring in USD offers a more stable and consistent value representation, making the relationship more apparent.


3. Increased Correlation Per Transaction Basis: The correlation coefficient increases significantly when both transaction volume and fees are adjusted on a per-transaction basis. This adjustment normalizes the data, reducing disparities in scale and providing a more accurate reflection of the relationship between the volume and cost efficiency of individual transactions.


4. Spread of Data Points: The scatter of the points suggests variability in how much fees increase relative to the volume. Some dense clustering at lower volumes with varied fee ranges suggests that fees can be quite volatile and don't always increase linearly with volume.


5. Outliers: There appear to be outliers, especially in the scatterplots involving USD values. These could be due to periods of high price volatility in Bitcoin, where the dollar value of transactions and fees can be disproportionately high compared to the number of Bitcoins transacted.

```python
def yearly_vol_gas_corr(volume_series,gas_series):
    vt_yearly_correlations = {}

    for year, group in volume_series.groupby(volume_series.index.year):
        #yearly_data = pivot_data[str(year)]  # Data for the current year
        vt_yearly_data = gas_series[gas_series.index.year == year]
        vt_yearly_correlations[year] = vt_yearly_data.corr(group)
    return vt_yearly_correlations
```
