## Hypothesis on Correlations Between BTC and Different Asset Classes' ETFS

- Bonds (BIL, BNDX, LQD, TIP, IEI, TLT): Cryptocurrencies like Bitcoin have shown periods of both positive and negative correlations with bonds. During times of market uncertainty or economic instability, some investors view cryptocurrencies as a hedge against traditional assets like bonds, resulting in potentially higher correlations. However, this correlation can vary based on market sentiment and risk perception.

- Commodities (COW, DBA, GLD, SLV, UNG, USO): Cryptocurrencies, particularly Bitcoin, have at times shown correlations with commodities like gold and silver. Some investors perceive cryptocurrencies as digital commodities or "digital gold," resulting in periods of positive correlation, especially during market turmoil or inflation concerns.

- Equities (FM, VWO, SCHB, VEA, VNQ): Cryptocurrencies have displayed varying correlations with global equities. During certain periods, they've shown a positive correlation, especially with emerging market stocks. However, this correlation can be less consistent due to the different market dynamics and regulatory environments in various regions.

- Fixed Income and High Yield (HYG, MBB, SHY): Cryptocurrencies have had mixed correlations with high-yield bonds and short-term treasuries. Their correlations with fixed income assets can fluctuate based on market sentiment, interest rate expectations, and macroeconomic factors.

- Diversified Portfolios (PSP): This ETF represents a diversified portfolio of global listed private equity. The correlation with cryptocurrencies might vary widely due to the nature of private equity investments, which can be less influenced by short-term market trends.

```python
# Your DataFrame structure
# Assuming 'data' is your DataFrame with columns: 'ticker', 'date', 'adj_close', 'adj_volume'

# Pivot the DataFrame to have each ETF as a column
pivot_data = etf_data.pivot(index='date', columns='ticker', values='adj_close')

# Display the pivoted DataFrame
pivot_data
```

```python
idx = bitcoin_price.index.intersection(pivot_data.index)
```

```python
bitcoin_price = bitcoin_price.loc[idx]
bitcoin_price = bitcoin_price.sort_index()
```

```python
correlations = pivot_data.corrwith(bitcoin_price['value'], axis=0)
```

```python
correlations.sort_values(ascending=False)
```

```python
bitcoin_price_series = bitcoin_price['value']
```

```python
bitcoin_price_series
```

```python
etf_data_indexed = etf_data.set_index('date')
```

```python
# Calculate yearly correlations
yearly_correlations = {}

for year, group in bitcoin_price_series.groupby(bitcoin_price_series.index.year):
    #yearly_data = pivot_data[str(year)]  # Data for the current year
    yearly_data = pivot_data[pivot_data.index.year == year]
    yearly_correlations[year] = yearly_data.corrwith(group)

# Create a DataFrame for yearly correlations
yearly_correlations_df = pd.DataFrame(yearly_correlations)

# Display the DataFrame with yearly correlations
print(yearly_correlations_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have the DataFrame 'yearly_correlations_df' containing yearly correlations

# Create a heatmap for yearly correlations
plt.figure(figsize=(10, 8))
sns.heatmap(yearly_correlations_df, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Yearly Correlations: Bitcoin vs ETFs')
plt.xlabel('ETFs')
plt.ylabel('Year')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Number of subplots to create
num_subplots = len(yearly_correlations_df.index)

# Create subplots for each ETF's yearly correlation trend with Bitcoin
fig, axes = plt.subplots(num_subplots, 1, figsize=(8, 6*num_subplots), sharex=True)

for i, ETF in enumerate(yearly_correlations_df.index):
    axes[i].plot(yearly_correlations_df.loc[ETF])
    axes[i].set_title(f'Yearly Correlation Trend: Bitcoin vs {ETF}')
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Correlation')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Loop through each ETF in the index and plot its yearly correlations
for ETF in bond_etfs:
    plt.plot(yearly_correlations_df.loc[ETF], label=ETF)

plt.xlabel('Year')
plt.ylabel('Correlation')
plt.title('Yearly Correlation Trends: Bitcoin vs ETFs')
plt.legend()
plt.grid(True)
plt.show()
```

## Volume Correlation with Transation Fee Core Findings:

1. Positive Correlation Between Volume and Fee: There is a clear positive correlation between transaction volume and transaction fees. This indicates that as more transactions occur within the Bitcoin network, the total transaction fees tend to increase.


2. Higher Correlation in USD Terms: The correlation is stronger when transaction volumes and fees are measured in USD rather than in raw Bitcoin. This can be attributed to the influence of Bitcoin's price volatility. When measuring in Bitcoin, the inherent price fluctuations can mask the underlying relationship between transaction volume and fees. In contrast, measuring in USD offers a more stable and consistent value representation, making the relationship more apparent.


3. Increased Correlation Per Transaction Basis: The correlation coefficient increases significantly when both transaction volume and fees are adjusted on a per-transaction basis. This adjustment normalizes the data, reducing disparities in scale and providing a more accurate reflection of the relationship between the volume and cost efficiency of individual transactions.


4. Spread of Data Points: The scatter of the points suggests variability in how much fees increase relative to the volume. Some dense clustering at lower volumes with varied fee ranges suggests that fees can be quite volatile and don't always increase linearly with volume.


5. Outliers: There appear to be outliers, especially in the scatterplots involving USD values. These could be due to periods of high price volatility in Bitcoin, where the dollar value of transactions and fees can be disproportionately high compared to the number of Bitcoins transacted.

$\large\textbf{8 }\quad\large\textbf{Interest Rate Scenario Analysis (Low vs High)}$

```python
import pandas_datareader as pdr

# Fetch the Federal Funds Rate
fed_funds_rate = pdr.get_data_fred('FEDFUNDS', start_date, end_date)
```

```python
fed_funds_rate
```

```python
#define low / high interest rates as 1st / 3rd quartile
fed_low = fed_funds_rate.quantile(0.25)['FEDFUNDS']
fed_high = fed_funds_rate.quantile(0.75)['FEDFUNDS']
print("low interest rate scenario as int rate <= ",fed_low)
print("high interest rate scenario as int rate >= ",fed_high)
```

```python
fed_funds_rate.plot(title="Fed Funds Rate over time")
plt.axhline(fed_low, color='r', linestyle='--', label='Low Interset Rate: 1st Quantile')
plt.axhline(fed_high, color='g', linestyle='--', label='High Interest Rate: 3rd Quantile')
plt.legend()
```

```python
low_interest_dates = fed_funds_rate[fed_funds_rate['FEDFUNDS']<=fed_low].index
low_int_etf_data = pivot_data[pivot_data.index.to_period('M').isin(low_interest_dates.to_period('M'))]
low_int_bitcoin_price = bitcoin_price_series[bitcoin_price_series.index.to_period('M').isin(low_interest_dates.to_period('M'))]

high_interest_dates = fed_funds_rate[fed_funds_rate['FEDFUNDS']>=fed_high].index
high_int_etf_data = pivot_data[pivot_data.index.to_period('M').isin(high_interest_dates.to_period('M'))]
high_int_bitcoin_price = bitcoin_price_series[bitcoin_price_series.index.to_period('M').isin(high_interest_dates.to_period('M'))]
```

```python
low_int_corr=low_int_etf_data.corrwith(low_int_bitcoin_price)
high_int_corr = high_int_etf_data.corrwith(high_int_bitcoin_price)
int_rate_correlation_df = pd.DataFrame(data={'low_interest_rate':low_int_corr,'high_interest_rate':high_int_corr})
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(int_rate_correlation_df, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('High vs Low Interest Rate Scenario Correlations: Bitcoin vs ETFs')
plt.xlabel('Interest Rate Scenario')
plt.ylabel('ETFs')
plt.show()
```

## Interest Rate Influence on Correlations between Bitcoin and Traditional Asset

- Cash (BIL):

    - In low-rate environments, negative correlation reflects investors' preference for higher-yielding assets like Bitcoin.
    - High rates drive a positive correlation as cash becomes more attractive, drawing funds away from volatile assets like Bitcoin.

- Equities (VEA, VWO, SCHB):

    - They typically move with Bitcoin in low-rate scenarios, encouraged by cheap capital and growth-seeking behavior.
    - In high-rate environments, the correlation can vary. Some equities remain positively correlated due to their specific market factors, while others may diverge as rising rates discourage risk-taking.

- Fixed Income (BNDX, EMB, TIP, TLT):

    - Shows less consistent correlation patterns. In low-rate periods, these may not provide attractive yields, leading investors to Bitcoin, while high rates can make these assets more appealing relative to Bitcoin.
        - TLT (Long-term Treasuries): Shows a strong negative correlation in both scenarios, emphasizing its role as a counterbalance to Bitcoin's volatility, more so in high rates due to increased yields.
        - TIP (Treasury Inflation-Protected Securities): Maintains a positive correlation in low rates, likely due to investors' concerns about inflation which can prompt a turn towards both TIPS and Bitcoin.

- Commodities (COW, DBA, GLD, SLV):

    - Often correlate with Bitcoin during low-rate periods as alternative investments, especially when inflation concerns drive investors towards "hard assets."
    - High rates can lead to negative correlations as commodities can be less favored due to costlier financing and lower industrial demand.
        - GLD (Gold) and SLV (Silver): Both maintain positive correlations in low rates but diverge in high rates; GLD's correlation becomes very strong, reinforcing its status as a safe haven alongside Bitcoin during high inflation or currency devaluation periods.
        - USO (Oil): Displays a sharp negative correlation in high-rate scenarios, possibly due to reduced economic activity affecting oil prices and contrasting investment flows 

- Private Equity (PSP):

    - Demonstrates a robust positive correlation with Bitcoin in low-rate scenarios due to the allure of higher returns from both asset classes.
    - Maintains a positive correlation even when rates rise, potentially due to long-term investment horizons aligning with Bitcoin's growth prospects.
    
- Real Estate (VNQ):
    - Exhibits a notably lower positive correlation in high rates compared to other equities, perhaps due to the sensitivity of real estate to interest rate hikes, which can dampen growth expectations.


Differences in correlations across interest rate scenarios reflect the varying risk appetites and investment strategies employed by market participants. Assets more sensitive to interest rates, like bonds and certain commodities, exhibit larger swings in correlation, while those with longer-term outlooks or less interest rate sensitivity, like private equity, show more stability.
